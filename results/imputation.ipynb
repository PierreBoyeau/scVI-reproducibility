```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def barplot(mean, alg, title, save=None):
    ind = np.arange(len(mean))  # the x locations for the groups
    width = 0.35       # the width of the bars
    fig = plt.figure(dpi=200)
    ax = fig.add_subplot(111)
    rects1 = ax.barh(ind, mean, width, color='r')
    #ax.axvline(x=np.min(np.array(mean)[np.array(mean) > 0]), ymin=0, ymax = 5, linewidth=1, color='b')

    # add some text for labels, title and axes ticks
    ax.set_xlabel(title)
    ax.set_yticks(ind)
    ax.set_yticklabels(alg)
    plt.tight_layout()
```

## ZEILER


```python
data = [ 9.6733, 6.969, 3.16874, 2.299, 2.325] #uniform
#data = [ 9.25, 4.97, 6, 2.299, 2.45]
```

```python
barplot(data, ["MAGIC", "BISCUIT","ZIFA", "ZINB-WaVE", "scVI"], 'Median Imputation error (CORTEX)')
plt.savefig("zeisel_imputation.pdf", dpi=300)
```
